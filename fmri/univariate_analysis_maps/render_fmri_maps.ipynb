# Plot Juror fMRI Results

```python
from nilearn import plotting, datasets, surface
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=300
```

```python
#crime type bias
crime_type_bias = 'univariate_z-stat_maps/juror_scenario_thresh.nii.gz'
img_ctb = crime_type_bias

#background anatomical
img_bg = 'underlay/mni152_2009_256.nii.gz'
```

# Crime Type Bias

## Right hemisphere

```python
big_fsaverage = datasets.fetch_surf_fsaverage('fsaverage')
big_texture = surface.vol_to_surf(img_ctb, big_fsaverage.pial_right, radius=0, interpolation='nearest')
#lateral view
ctb_right_lateral = plotting.plot_surf_stat_map(big_fsaverage.infl_right,
                                                big_texture,
                                                hemi='right',
                                                view = 'lateral',
                                                colorbar=False,
                                                threshold=2.3,
                                                vmax=6.0,
                                                bg_map=big_fsaverage.sulc_right,
                                                output_file = 'figs/rendered_brain_ctb_right_hem_lateral.png')
#medial view
ctb_right_medial = plotting.plot_surf_stat_map(big_fsaverage.infl_right,
                                               big_texture,
                                               hemi='right',
                                               view = 'medial',
                                               colorbar=False,
                                               threshold=2.3,
                                               vmax=6.0,
                                               bg_map=big_fsaverage.sulc_right,
                                               output_file = 'figs/rendered_brain_ctb_right_hem_medial.png')
```
